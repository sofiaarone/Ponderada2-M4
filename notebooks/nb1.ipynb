### Gráfico de carga em C

```python
# Processar os dados para extrair tempo e carga no capacitor (segunda coluna)
linhas = dados.strip().split('\n')
tempo = []
carga = []

for linha in linhas:
    partes = linha.split()
    tempo.append(int(partes[0]))
    carga.append(float(partes[1]))  # Tensão do capacitor

# Plotar o gráfico da carga no capacitor
plt.figure(figsize=(8,5))
plt.plot(tempo, carga, marker='o', linestyle='-', color='b')
plt.title('Gráfico da carga no capacitor')
plt.xlabel('Tempo (ms)')
plt.ylabel('Tensão (V)')
plt.grid(True)
plt.show()

```

### Gráfico de descarga em R

```python
# Processar os dados para extrair tempo e descarga no resistor (terceira coluna)
linhas = dados.strip().split('\n')
tempo = []
descarga = []

for linha in linhas:
    partes = linha.split()
    tempo.append(int(partes[0]))
    descarga.append(float(partes[2]))  # Tensão de descarga no resistor

# Plotar o gráfico da descarga no resistor
plt.figure(figsize=(8,5))
plt.plot(tempo, descarga, marker='o', linestyle='-', color='r')
plt.title('Gráfico da descarga no resistor')
plt.xlabel('Tempo (ms)')
plt.ylabel('Tensão (V)')
plt.grid(True)
plt.show()
```

### Gráfico de comparação

```python
# Processando dados
linhas = dados.strip().split('\n')
tempo = []
carga = []
descarga = []

for linha in linhas:
    t, v_carga, v_descarga = linha.split()
    tempo.append(int(t))
    carga.append(float(v_carga))
    descarga.append(float(v_descarga))

# Plotando os dados em um único gráfico para comparação
plt.figure(figsize=(10,6))
plt.plot(tempo, carga, label='Carga no Capacitor (V)', color='blue', marker='o')
plt.plot(tempo, descarga, label='Descarga no Resistor (V)', color='red', marker='o')
plt.title('Comparação da Carga e Descarga no Circuito RC')
plt.xlabel('Tempo (ms)')
plt.ylabel('Tensão (V)')
plt.legend()
plt.grid(True)
plt.show()
```
